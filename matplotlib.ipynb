```python
df = pdf.read_csv('Used_Bikes.csv')
df
```

```python
df.drop_duplicates(inplace=True)
```

```python
df['brand'].value_counts()
```

```python
brand_name = list(df['brand'].value_counts().keys())
brand_name
```

```python
no_of_counts = list(df['brand'].value_counts().values)
no_of_counts
```

```python
plt.figure(figsize=(10,5)) ##width,height
plt.bar(brand_name,no_of_counts,color='red',ec='black')
plt.xticks(rotation=90)
plt.savefig('branddata')
plt.show()
```

```python
owner_count = list(df['owner'].value_counts())
brand_name = list(df['brand'].value_counts().keys())
```

```python
plt.figure(figsize=(10,5)) ##width,height
plt.bar(brand_name,owner_count,color='red',ec='black')
plt.xticks(rotation=90)
#plt.savefig('branddata')
plt.show()
```

```python
country_population = [450,730,900,450,800]
country_name = ['Japan','bangladesh','india','nepal','china']


plt.pie(country_population,labels=country_name)
plt.show()
```

```python
plt.figure(figsize=(10,6))
explosion = [0.2,0,0,0,0.35]
plt.pie(country_population,labels=country_name,autopct='%1.2f%%',explode=explosion,wedgeprops={'ec':'black','lw':3},shadow=True)
plt.savefig('pop-pie')
plt.show()
```
